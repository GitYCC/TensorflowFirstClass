# Python/Tensorflow神經網路＆深度學習入門

## 00_安裝教學

### MacOS

#### Step 1:  安裝 Xcode
打開你的App Store，搜尋Xcode並安裝。

#### Step 2:  下載Anaconda MacOS版
到 https://www.anaconda.com/download/#macos 頁面，下載「Python 3.6 version」，成功的話會有一個pkg檔被下載。

#### Step 3:  安裝Anaconda
點開這個pkg檔，然後無腦按「繼續」就可以安裝成功了，安裝完畢App會多一個`Anaconda-Navigator.app`。

#### Step 4:  從Github下載教材


#### Step 5:   設置虛擬環境
繼續留在「終端機」，首先先更新一下`conda`
```
$ conda update conda
```

接下來創造一個新的「虛擬環境」，並且進入這個「虛擬環境」
```
$ conda create -n myLab python=3 anaconda
$ source activate myLab
```

如果你想要離開「虛擬環境」
```
$ source deactivate
```

如果你想要刪除「虛擬環境」
```
$ conda remove -n myLab -all
```


#### Step 6:  安裝Python套件 

接下來，我們就在「虛擬環境」裡面安裝我們未來課程需要用的套件。首先先增加額外的channel來確保我們可以搜尋的到tensorflow。
```
(myLab) $ conda config --append channels conda-forge
```


接下來，直接使用我教材中提供的`requirements.txt`來安裝。
```
(myLab) $ conda install --yes --file requirements.txt
```

或者，如果你想要自己一個一個安裝，體驗安裝的感覺！？也可以這樣做
```
(myLab) $ conda install msgpack==0.5.6
(myLab) $ conda install jupyter==1.0.0
(myLab) $ conda install numpy==1.14.3
(myLab) $ conda install pandas==0.22.0
(myLab) $ conda install matplotlib==2.2.2
(myLab) $ conda install scipy==1.1.0
(myLab) $ conda install scikit-learn==0.19.1

(myLab) $ conda install tensorflow==1.8.0
(myLab) $ conda install tensorflow-gpu==1.8.0 # if you want GPU support

(myLab) $ conda install bleach==1.5.0 
(myLab) $ conda install Keras==2.1.6 
(myLab) $ conda install imageio==2.3.0
(myLab) $ conda install Pillow==5.1.0
(myLab) $ conda install scikit-image==0.13.1
```

這兩種作法是等價的。

#### Step 7:  啟動 Jupyter Notebook
```
(myLab) $ jupyter-notebook
```
