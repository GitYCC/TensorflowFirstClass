# Python/Tensorflow神經網路＆深度學習入門

## 安裝教學

### Python2.7 安裝 on MacOS

#### Step 1:  安裝 Xcode
打開你的App Store，搜尋Xcode並安裝。

#### Step 2: 安裝Homebrew
安裝 [Homebrew](https://brew.sh) 這個Mac上好用的套件管理，打開「終端機」，輸入
```
$ /usr/bin/ruby -e "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/master/install)"
```
安裝完畢你就可以直接在「終端機」上使用它，我們試著搜尋「python」關鍵字看有沒有相應套件
```
$ brew update
$ brew search python
```
有的！

#### Step 3: 安裝python2.7
安裝python2.7
```
$ brew install python
```

安裝完畢後檔案會被放在底下這個路徑，你可以打開來看一下
```
$ open /usr/local/Cellar
```

應該就會看到python的資料夾了。

#### Step 4:  設定路徑 `$PATH`（不跟系統 Python 打架）

這是什麼呢？當你輸入`brew` , `open` , `python2.7` 這些指令到「終端機」上，為什麼「終端機」會認的了這些指令，原因就出在於這個`PATH`上，又稱為「環境變數」，我們把它叫出來看看
```
$ echo $PATH
/usr/local/bin:/usr/bin:/bin:/usr/sbin:/sbin
```
你可以看到有五個路徑分別被字元 : 隔開，由前到後分別為`/user/local/bin`、`/usr/bin`、`/bin`、`/usr/sbin`、`/sbin`，這一些都是裝有執行檔的資料夾，今天你如果輸入某個指令，他就會從第一個資料夾下面開始找起，也就是`/user/local/bin`，沒有找到再依序往下找，直到找不到為止，如果今天`/usr/bin`底下有python，而你剛剛用`brew`安裝的另一個python放在`/user/local/bin` 底下，在這個例子中，你會執行到的就是第一個路徑`/user/local/bin` 下的python，那這也是我們要的結果，我們想要執行我們自己安裝的，而不是系統原有的。

如果`/user/local/bin`不是在第一個的話，就必須去修改PATH的順序。
```
$ sudo emacs /etc/paths
```
輸入密碼後，就會進入修改模式，然後開始修改順序，利用以下指令把/user/local/bin 放到最上面

control + k：把一行字剪下來

control + y：把字貼上

control + x + s：存檔

control + x + c：關掉 emacs

修改完成重開「終端機」，讓環境變數重載，在輸入一次 echo $PATH 應該就可以看到修改後正確的環境變數了。

#### Step 5: 最後檢查
那就安裝完畢啦！最後檢查一下你下python2.7的時候是不是來自於`/user/local/bin`
```
$ which python2.7
/usr/local/bin/python2.7
```

### Python套件安裝 on MacOS / Linux
#### Step 1: 
安裝pip
```
$ curl "https://bootstrap.pypa.io/get-pip.py" -o "get-pip.py"
$ sudo python2.7 get-pip.py
``` 

#### Step 2: 
安裝套件
```
$ sudo pip2.7 install --upgrade pip
$ sudo pip2.7 install jupyter==1.0.0
$ sudo pip2.7 install numpy==1.14.3
$ sudo pip2.7 install pandas==0.22.0
$ sudo pip2.7 install matplotlib==2.2.2
$ sudo pip2.7 install tensorflow==1.8.0 (or tensorflow-gpu==1.8.0 if you want GPU support)
$ sudo pip2.7 install scikit-learn==0.19.1
``` 

或者你想要快速安裝，也可以下載目錄底下的`requirements.txt`檔案，並執行
```
$ sudo pip2.7 install --upgrade pip
$ sudo pip2.7 install --upgrade -r requirements.txt
```

就可以直接安裝全部的相關套件。

### Python2.7 安裝 on Linux
#### Step 1: 先安裝一些相依套件
```
$ sudo apt-get update
$ sudo apt-get install build-essential checkinstall
$ sudo apt-get install libreadline-gplv2-dev libncursesw5-dev libssl-dev libsqlite3-dev tk-dev libgdbm-dev libc6-dev libbz2-dev
```

#### Step 2: 從網路上下載python2.7 source code
```
$ cd ~/Downloads
$ wget https://www.python.org/ftp/python/2.7.13/Python-2.7.13.tgz
```

#### Step 3: 解壓縮並進去資料夾
```
$ tar xzf Python-2.7.13.tgz
$ cd Python-2.7.13
```

#### Step 4: 依環境配置並安裝
```
$ sudo ./configure
$ sudo make altinstall
```

make altinstall 是為了避免你去取代掉預設的python在/usr/bin/python。
